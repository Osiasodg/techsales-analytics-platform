```python
df['campagne_marketing'] = df['campagne_marketing'].fillna('Non renseigné')

# Vérification
print(f"Valeurs manquantes restantes : {df['campagne_marketing'].isnull().sum()}")
print(f"\nDistribution campagne_marketing :")
print(df['campagne_marketing'].value_counts())
```

```python
# Vérification des valeurs aberrantes sur les colonnes numériques clés
print("=" * 50)
print("VALEURS MIN ET MAX PAR COLONNE NUMÉRIQUE")
print("=" * 50)

colonnes = ['age_client', 'montant_brut_eur', 'montant_net_eur', 
            'remise_pct', 'score_satisfaction', 'nps_score', 
            'taux_utilisation_pct', 'nb_tickets_support', 'uptime_pct']

for col in colonnes:
    print(f"{col:30} | Min: {df[col].min():10.2f} | Max: {df[col].max():10.2f} | Moyenne: {df[col].mean():10.2f}")
```

```python
# Convertir la date en format datetime
df['date_transaction'] = pd.to_datetime(df['date_transaction'])

# Extraire mois, année et trimestre
df['mois'] = df['date_transaction'].dt.month
df['annee'] = df['date_transaction'].dt.year
df['trimestre'] = df['date_transaction'].dt.quarter

# Créer la colonne saison
def get_saison(mois):
    if mois in [12, 1, 2]:
        return 'Hiver'
    elif mois in [3, 4, 5]:
        return 'Printemps'
    elif mois in [6, 7, 8]:
        return 'Eté'
    else:
        return 'Automne'

df['saison'] = df['mois'].apply(get_saison)

# Créer la colonne marge nette
df['marge_nette'] = df['montant_net_eur'] - df['cout_acquisition_eur']

# Créer la colonne ROI
df['roi_acquisition'] = (df['montant_net_eur'] / df['cout_acquisition_eur']).round(2)

# Vérification
print(" Nouvelles colonnes créées !")
print(f"\nAperçu des nouvelles colonnes :")
df[['date_transaction', 'mois', 'annee', 'trimestre', 'saison', 'marge_nette', 'roi_acquisition']].head(10)
```

```python
# Sauvegarder le dataset propre
df.to_csv('../data/processed/dataset_nettoye.csv', index=False)
print("Dataset nettoyé sauvegardé dans data/processed/")
print(f"Dimensions finales : {df.shape[0]} lignes × {df.shape[1]} colonnes")
```
